# Basic Plotting , Axes Lalbels and Title

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# plt.style.use(['dark_background'])
# %matplotlib inline
```

```python
## Ploting two 1-D Array
x = np.linspace(5, 100, 100)
y = np.linspace(10, 1000, 100)
plt.plot(x, y)
plt.show()
```

```python
# can work with list also, though it convertsit into np arrays internally
plt.plot([1,2,3,4],[5,6,7,8])
plt.show()
```

# Puting x and y labels and chart titles

```python
plt.plot(x,y)
plt.xlabel("current")
plt.ylabel("voltage")
plt.title("ohms law")
plt.show()
```

## Define the range of labels of the axis

```python
plt.plot(x,y)
plt.xlim([20, 800])
plt.ylim([200, 800])
plt.show()
```

# Change the colors and line type

```python
x = np.linspace(0,10,20)
y = x*2
```

```python
# Color blue with line type '+'
plt.plot(x,y, 'b*')
plt.show()
```

# Plotting multiple plots in same plot

```python
x= np.linspace(0,5,10)
y= np.linspace(3,6,10)
```

## Plot three curves, y, y**2 and y**3 with difeerent line

```python
plt.plot(x,y, 'r-', x, y**2, 'b+', x, y**3, 'g^')
plt.show()
```

# Figures and SubPlots

`Often we need to create multiple plots inthe same figure.
matplotlib has the concept of figures and subplots using which we can create multiple subplots inside the same figure
to crate multiple subplots in the same figure, we can use the plt.subplot(nrows, ncols, nsubplot)`

```python
x = np.linspace(1,10, 100)
y= np.log(x)
```

```python
# Instatiate a new figure explicity
plt.figure(1)
```

```python
# Create a subplot with 1 row , 2 column
# Create the first subplot in figure 1
plt.subplot(121)
plt.title("y=log(x)")
plt.plot(x,y)

# Create the first subplot in figure 1
plt.subplot(122)
plt.title("y=log(x)**2")
plt.plot(x,y**2)
plt.show()
```

## Lets see another example - 4 subplots in two rows and two columns

```python
x = np.linspace(1,10,100)
plt.figure(1) # OPTIONAL COMMAND

plt.subplot(2,2,1)
plt.title("Linear")
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title("square")
plt.plot(x,x**2)

plt.subplot(2,2,3)
plt.title("Cubic")
plt.plot(x,x**3)

plt.subplot(2,2,4)
plt.title("Square root")
plt.plot(x,np.sqrt(x))

plt.tight_layout()
plt.show()
```

# Types of Commonly used Plots

## Lets use retail stores sales data to create some commonly use plots
1. BoxPlot
2. Histograms
3. Scatter Plot
4. Bar plots

```python
df = pd.read_csv("market_fact.csv")
df.head()
```

## BoxPlot

```python
# BoxPlot: Visualize the distribution of a continous varibale
plt.boxplot(df["Order_Quantity"])
plt.show()
```

```python
plt.boxplot(df.Sales)
plt.show()
```

```python
## Scale the data for better visualization when the variace is so high
plt.boxplot(df.Sales)
plt.yscale("log")
plt.show()
```

## Histogram

```python
plt.hist(df.Sales)
plt.show()
```

```python
## for higher variance in data log the y scale
plt.hist(df.Sales)
plt.yscale("log")
plt.show()
```

## Scatter Plot

```python
## Scatter plot with two variables
plt.scatter(df.Sales, df.Profit, color='red')
plt.show()
```
